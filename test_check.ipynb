```python
import pandas as pd
data = pd.read_csv('test_results.csv',header=None)
```

```python
original_data = data.loc[:,1].to_list()
test_data = data.loc[:,2].to_list()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,np.max(original_data),100)
y = x
plt.plot(original_data,test_data,'.')
plt.plot(x,y)
plt.xlabel('original Tc')
plt.ylabel('pred Tc')
plt.savefig('regression_mag_graph.png')
```

```python
# accuracy check
import pandas as pd
data = pd.read_csv('test_results.csv',header=None)
```

```python
Y_test = data.iloc[:,1]
Y_score = data.iloc[:,2]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,recall_score,f1_score
from sklearn.metrics import roc_curve, precision_recall_curve, auc
 
fpr, tpr, thresholds = roc_curve(y_true=Y_test,y_score=Y_score)
plt.plot(fpr,tpr,label='roc curve (AUC = %0.3f)' % auc(fpr,tpr))
plt.plot([0,0,1], [0,1,1], linestyle='--', label='ideal line')
plt.plot([0, 1], [0, 1], linestyle='--', label='random prediction')
plt.legend()
plt.xlabel('false positive rate(FPR)')
plt.ylabel('true positive rate(TPR)')
# plt.show()
plt.savefig('roc_class_mag.png')
```
